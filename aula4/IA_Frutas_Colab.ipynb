# 🌟 IA Simples com Python: Classificação de Frutas

Este notebook demonstra como criar uma **Inteligência Artificial simples** usando **árvore de decisão** para classificar frutas com base em **peso** e **cor**. Ideal para fins educacionais e introdução à IA com Python.


```python
# ✅ Instalar bibliotecas necessárias
!pip install scikit-learn pandas matplotlib --quiet

# 📚 Importar bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# 🍎 Criar conjunto de dados simples
# Cada fruta tem um peso (em gramas) e uma cor (0 = verde, 1 = vermelho)

dados = {
    'peso': [150, 180, 120, 130, 200, 170],
    'cor': [0, 0, 1, 1, 0, 1],
    'fruta': ['maçã', 'maçã', 'morango', 'morango', 'maçã', 'morango']
}

df = pd.DataFrame(dados)
df
```

```python
# 🔄 Preparar os dados para treino e teste

X = df[['peso', 'cor']]  # Atributos
y = df['fruta']          # Rótulo

# Dividir em treino (70%) e teste (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 🧠 Criar e treinar o modelo de árvore de decisão

modelo = DecisionTreeClassifier()
modelo.fit(X_train, y_train)
```

```python
# 📊 Avaliar o modelo

y_pred = modelo.predict(X_test)
print("Acurácia do modelo:", accuracy_score(y_test, y_pred))
```

```python
# 🔍 Fazer uma previsão
# Exemplo: fruta com peso 160g e cor verde (0)

nova_fruta = [[160, 0]]
print("Previsão:", modelo.predict(nova_fruta))
```

```python
# 🌳 Visualizar a árvore de decisão

plt.figure(figsize=(10,6))
tree.plot_tree(modelo, feature_names=['peso','cor'], class_names=modelo.classes_, filled=True)
plt.title("Árvore de Decisão - Classificação de Frutas")
plt.show()
```
